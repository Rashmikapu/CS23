- During training, the classifier takes the training data and simply remembers it

We can also use other distance metrics such as L1 distance.
For pixel values $p_{ij}^{(k)}$ at location $(i,j)$ of some image $I_k$,

the mean $\mu$ across all pixels over all images is $$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$
And the pixel-wise mean $\mu_{ij}$ across all images is
$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$
The general standard deviation $\sigma$ and pixel-wise standard deviation $\sigma_{ij}$ is defined similarly.

Which of the following preprocessing steps will not change the performance of a Nearest Neighbor classifier that uses L1 distance? Select all that apply.
1. Subtracting the mean $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. Subtracting the per pixel mean $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$.
4. Subtracting the pixel-wise mean $\mu_{ij}$ and dividing by the pixel-wise standard deviation $\sigma_{ij}$.
5. Rotating the coordinate axes of the data.

$\color{blue}{\textit Your Answer:}$ Options (1), (2), (3) will not change the performance of an NN classifier that uses L1 distance.


$\color{blue}{\textit Your Explanation:}$
 L1 distance before:

  L1 = |p$_{x}$ - p$_{x}$'|


 where p$_{x}$ corresponds to pixel of the train image at x location and p$_{x}$' corresponds to test image pixel at x location.

- **Option 1 : Subtracting the mean $\mu$:**
  
  L1 distance after subtracting $\mu$:

 L1 = |p$_{x}$ - $\mu$ - (p$_{x}$'-$\mu$)|

 The $\mu$ gets cancelled out and we are left with the same calculation of L1 as before. Hence, L1 distance remains unaffected by this operation.

- **Option 2 : Subtracting per pixel mean $\mu_{ij}$:**

 L1 distance after subtracting $\mu$:

   L1 = |p$_{x}$ - $\mu_{x}$ - (p$_{x}$'-$\mu_{x}$)|

  where $\mu_{x}$ is the corresponding per pixel mean value.
 The $\mu_{x}$ gets cancelled out and we are left with the same calculation of L1 as before. Hence, L1 distance remains unaffected by this operation.

- **Option 3 : Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$ :**

  L1 distance after Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$:

  L1 = |$\frac{p_{x} - \mu - (p_{x}'-\mu)}{\sigma}| = |\frac{p_{x} - p_{x}'}{\sigma}$|

  We see that the original L1 distance is being scaled by a factor of ${\sigma}$. Since this value does not vary with every pixel, it remains a constant scaling factor for distance for every pixel. Hence, the performance is unaffected since the effect remains unchanged.

- **Option 4 : Subtracting the pixel-wise mean $\mu_{x}$ and dividing by the pixel-wise standard deviation $\sigma_{x}$ :**

 L1 distance after subtracting the pixel-wise mean $\mu_{x}$ and dividing by the pixel-wise standard deviation $\sigma_{x}$ :

 L1 = |$\frac{p_{x} - \mu_{x} - (p_{x}'-\mu_{x})}{\sigma}| = |\frac{p_{x} - p_{x}'}{\sigma_{x}}$|

 The value of $\sigma_{x}$ varies for every pixel because it is calculated as the deviation from the per pixel mean. Therefore, it changes the distance values based on the standard deviation which is not constant, unlike the previous case. Hence this operation affects the performance of the classifier since distance metric is altered.

- **Option 5 : Rotating the coordinate axes of the data :**
 The rotation matrix is [$\begin{smallmatrix} Cos\theta & -Sin\theta \\ Sin\theta & Cos\theta\end{smallmatrix}$]
 Let us assume the axes are rotated -45$^0$, then

  Rotation matrix = [$\begin{smallmatrix} 0.85 & -0.85 \\ 0.85 & 0.85\end{smallmatrix}$] (approximately)


  Suppose $p_{x}$ = [0 , 1]
  and $p_{x}^1$ = [1,0]
  
  L1 before rotation = 1+1 = 2

  After rotation,
  
  new $p_{x}$ = [0 , 1][$\begin{smallmatrix} 0.85 & -0.85 \\ 0.85 & 0.85\end{smallmatrix}$] = [0.85 , 0.85]

  new $p_{x}^1$ = [1 , 0][$\begin{smallmatrix} 0.85 & -0.85 \\ 0.85 & 0.85\end{smallmatrix}$] = [0.85 , -0.85]

  L1 after rotation = 0+0.85$^2$ = 0.7225

  Hence, the change in distance metric might affect the performance of the classifier.

### Cross-validation

We have implemented the k-Nearest Neighbor classifier but we set the value k = 5 arbitrarily. We will now determine the best value of this hyperparameter with cross-validation.

```````````````````````````````````````````````python
# PLEASE DO NOT MODIFY THE MARKERS
print('||||||||||||||||||||||||||||||||||||||||||||||')
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

X_train_folds = []
y_train_folds = []
################################################################################
# TODO:                                                                        #
# Split up the training data into folds. After splitting, X_train_folds and    #
# y_train_folds should each be lists of length num_folds, where                #
# y_train_folds[i] is the label vector for the points in X_train_folds[i].     #
# Hint: Look up the numpy array_split function.                                #
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

#np.split splits the data into the specified number of subarrays and returns an
#array of arrays. We split training data into given num_folds (5 folds in this
# case) using this function.
X_train_folds = np.split(X_train, num_folds)
y_train_folds = np.split(y_train, num_folds)
# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# A dictionary holding the accuracies for different values of k that we find
# when running cross-validation. After running cross-validation,
# k_to_accuracies[k] should be a list of length num_folds giving the different
# accuracy values that we found when using that value of k.
k_to_accuracies = {}


################################################################################
# TODO:                                                                        #
# Perform k-fold cross validation to find the best value of k. For each        #
# possible value of k, run the k-nearest-neighbor algorithm num_folds times,   #
# where in each case you use all but one of the folds as training data and the #
# last fold as a validation set. Store the accuracies for all fold and all     #
# values of k in the k_to_accuracies dictionary.                               #
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# loop over every k value and create a list for key = k in k_to_accuracies dictionary
# to store accuracy for each fold

for k in k_choices :
    k_to_accuracies[k] = []
    # create train and validation datasets for every fold by assigning ith fold
    # for validation and concatenating rest of the folds for training
    for i in range(num_folds):
      X_train_current = np.concatenate([X_train_folds[j] for j in range(num_folds) if j != i])
      y_train_current = np.concatenate([y_train_folds[j] for j in range(num_folds) if j != i])
      val_x = X_train_folds[i]
      val_y = y_train_folds[i]
      #train the classifier for the current k value and fold
      classifier.train(X_train_current, y_train_current)
      validation_pred = classifier.predict(val_x, k ,0 )
      count = 0
      # count the no.of correct predictions
      for pred in range(len(validation_pred)) :
        if validation_pred[pred] == val_y[pred] :
          count+=1
      # append the accuracy to the list in the dictionary for key = k
      k_to_accuracies[k].append(count/len(val_y))
# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# Print out the computed accuracies
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print('k = %d, accuracy = %f' % (k, accuracy))
# PLEASE DO NOT MODIFY THE MARKERS
print('``````````````````````````````````````````````')
```````````````````````````````````````````````

```````````````````````````````````````````````python
# PLEASE DO NOT MODIFY THE MARKERS
print('||||||||||||||||||||||||||||||||||||||||||||||')
# plot the raw observations
for k in k_choices:
    accuracies = k_to_accuracies[k]
    plt.scatter([k] * len(accuracies), accuracies)

# plot the trend line with error bars that correspond to standard deviation
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation on k')
plt.xlabel('k')
plt.ylabel('Cross-validation accuracy')
plt.show()
# PLEASE DO NOT MODIFY THE MARKERS
print('``````````````````````````````````````````````')
```````````````````````````````````````````````

```````````````````````````````````````````````python
# PLEASE DO NOT MODIFY THE MARKERS
print('||||||||||||||||||||||||||||||||||||||||||||||')
# Based on the cross-validation results above, choose the best value for k,
# retrain the classifier using all the training data, and test it on the test
# data. You should be able to get above 28% accuracy on the test data.

# In the previous block, we found mean of accuracies for each k. Now we assign
# the k associated with the maximum value of average accuracy as the best_k
best_acc = accuracies_mean.argmax()
best_k = k_choices[best_acc]
classifier = KNearestNeighbor()
classifier.train(X_train, y_train)
y_test_pred = classifier.predict(X_test, k=best_k)

# Compute and display the accuracy
num_correct = np.sum(y_test_pred == y_test)
accuracy = float(num_correct) / num_test
print('Got %d / %d correct => accuracy: %f' % (num_correct, num_test, accuracy))
# PLEASE DO NOT MODIFY THE MARKERS
print('``````````````````````````````````````````````')
print(best_k)
```````````````````````````````````````````````

**Inline Question 3**

Which of the following statements about $k$-Nearest Neighbor ($k$-NN) are true in a classification setting, and for all $k$? Select all that apply.
1. The decision boundary of the k-NN classifier is linear.
2. The training error of a 1-NN will always be lower than or equal to that of 5-NN.
3. The test error of a 1-NN will always be lower than that of a 5-NN.
4. The time needed to classify a test example with the k-NN classifier grows with the size of the training set.
5. None of the above.

$\color{blue}{\textit Your Answer:}$ Options (2), (4) are true.


$\color{blue}{\textit Your Explanation:}$

- Option (1) : False because the decision boundary of KNN is not necessarily linear. Since the logic behind KNN is to predict the class labels by mapping the labels of majority vote among the k nearest neighbors, the distribution depends on the training data and their respective classes and need not be linear boundaries.

- Option (2) : True because the 1-NN classifier is already trained on training images. If we test on one such training image, the classifier maps it to the first nearest neighbor's class, which is the train image itsef. Hence, the prediction is accurate and training error is 0, which can be lower or equal to the 5-NN classifier, but not greater.

- Option (3) : False because the test error of 1-NN need not be always lower than that of a 5-NN. Since the 1-NN classifier does not take a majority vote among nearest neighbors but takes the class of first nearest neighbor as the prediction, there are high chances that the image can be misclassified by 1-NN compared to 5-NN. Having said that, there can also be a few cases wherein the images are rightly classified by 1-NN but not 5-NN, because 5-NN takes a majority vote and might wrongly predict the test label. Hence, it depends on the data and it cannot be said that the test error is ALWAYS higher for 1-NN than 5-NN.

- Option (4) : True because as the size of training data increases, more no.of distance values need to be calculated because distance matrix consists of distances between every train image and every test image. Hence, the computational time is increased and time taken for the prediction is higher.
